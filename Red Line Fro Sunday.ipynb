```python
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates


# Sort the arrival_times list
arrival_times.sort()

# Function to calculate time difference in minutes and round to the nearest minute
def time_difference(time1, time2):
    fmt = '%H:%M:%S'
    tdelta = datetime.strptime(time2, fmt) - datetime.strptime(time1, fmt)
    return round(tdelta.total_seconds() / 60)

# Prepare the data for the DataFrame
data = {
    'train number': range(1, len(arrival_times) + 1),
    'arrival time': arrival_times,
    'next train at': [time_difference(arrival_times[i], arrival_times[i+1]) 
                      if i < len(arrival_times) - 1 else 'N/A' 
                      for i in range(len(arrival_times))]
}

# Create the DataFrame
df = pd.DataFrame(data)



# Convert arrival times to datetime for plotting
arrival_times_dt = [datetime.strptime(time, '%H:%M:%S') for time in arrival_times]

# Filter out 'N/A' values for plotting
df_filtered = df[df['next train at'] != 'N/A']

plt.figure(figsize=(12, 6))

# Plotting the data
plt.plot(arrival_times_dt[:-1], df_filtered['next train at'], marker='o')

# Formatting the x-axis with specific time frames
plt.gca().xaxis.set_major_locator(mdates.HourLocator(interval=1))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%H:%M'))

plt.xlabel('Arrival Time')
plt.ylabel('Next Train At (Frequency in Minutes)')
plt.title('L.B.Nagar to Miyapur Sunday Frequency')
plt.xticks(rotation=45)
plt.grid(True)
plt.savefig('train_frequency_plot11.png', dpi=300, bbox_inches='tight')
plt.show()
```
